Problem - 3: Array Operations:

For the following arrays:

x = np.array([[1,2],[3,5]]) and y = np.array([[5,6],[7,8]]);

v = np.array([9,10]) and w = np.array([11,12]);

Complete all the task using numpy:

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# Task 1: Add the two arrays
addition = x + y
print("Addition of x and y:")
print(addition)

# Task 2: Subtract the two arrays
subtraction = x - y
print("\nSubtraction of x and y:")
print(subtraction)

# Task 3: Multiply the array x with an integer (e.g., 2)
multiplication = x * 2
print("\nMultiplication of x by 2:")
print(multiplication)

# Task 4: Find the square of each element in x
square = np.square(x)
print("\nSquare of each element in x:")
print(square)

# Task 5: Dot product
# Dot product between v and w
dot_vw = np.dot(v, w)
print("\nDot product of v and w:")
print(dot_vw)

# Dot product between x and v
dot_xv = np.dot(x, v)
print("\nDot product of x and v:")
print(dot_xv)

# Dot product between x and y
dot_xy = np.dot(x, y)
print("\nDot product of x and y:")
print(dot_xy)

# Task 6: Concatenate x and y along rows; Concatenate v and w along columns
concat_xy_row = np.concatenate((x, y), axis=0)  # Concatenating along rows
concat_vw_col = np.vstack((v, w))  # Concatenating along columns
print("\nConcatenation of x and y along rows:")
print(concat_xy_row)
print("\nConcatenation of v and w along columns:")
print(concat_vw_col)

# Task 7: Concatenate x and v
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("\nConcatenation of x and v:")
    print(concat_xv)
except ValueError as e:
    print("\nError during concatenation of x and v:")
    print(e)
```

Problem - 4: Matrix Operations:

• For the following arrays:

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:

```python
#1. Prove A.A−1 = I.
import numpy as np

# Define matrices A and B
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# Compute A^-1
A_inv = np.linalg.inv(A)

# Compute A * A^-1
result = np.dot(A, A_inv)
print("A * A^-1 = I (Identity Matrix):")
print(result)
```

```python
#2. Prove AB ̸= BA.
# Compute AB and BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nAB:")
print(AB)
print("BA:")
print(BA)
print("Is AB equal to BA? ", np.array_equal(AB, BA))
```

```python
#3. Prove (AB)transpose = B transpose A transpose
# Compute (AB)^T
AB_transpose = np.transpose(AB)

# Compute B^T * A^T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T:")
print(AB_transpose)
print("B^T * A^T:")
print(BT_AT)
print("Is (AB)^T equal to B^T * A^T? ", np.array_equal(AB_transpose, BT_AT))
```

```python
# Define coefficient matrix A and constant matrix B
A_eq = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Solve for X using np.linalg.inv
A_inv_eq = np.linalg.inv(A_eq)
X = np.dot(A_inv_eq, B_eq)

print("\nSolution of the system of equations using inverse method:")
print("X =", X)
```
