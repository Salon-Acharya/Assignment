Example 1: We will start with a simple problem. Consider the following linear systems:

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y1 = (5 - x) / 3
y2 = (5 + x) / 2
y3 = 3 - x

fig, ax = plt.subplots()

ax.plot(x, y1, color='r', label='$x+3y = 5$')
ax.plot(x, y2, color='g', label='$x-2y = -5$')
ax.plot(x, y3, color='b', label='$x+y = 3$')

ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)

ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)

ax.set_xticks(np.arange(-4, 5))
ax.set_aspect('equal')
ax.grid(True, ls=':')

plt.xlabel("X")
plt.ylabel("Y")

ax.legend()
plt.show()
```

**Task 1:**

Plot the graph for linear system b and write the point of intersection between three lines.

```python
from sympy import symbols, Eq, solve

# Define variables
x, y = symbols('x y')

# Define equations
eq1 = Eq(x + 3*y, 5)
eq2 = Eq(x - 2*y, -5)
eq3 = Eq(x + y, 3)

# Solve the system of equations
intersection = solve((eq1, eq2, eq3), (x, y))

print("Intersection point(s):", intersection)
import numpy as np
import matplotlib.pyplot as plt

# Define the equations
x = np.linspace(-5, 5, 100)
y1 = (5 - x) / 3
y2 = (5 + x) / 2
y3 = 3 - x

fig, ax = plt.subplots()

# Plot the equations
ax.plot(x, y1, color='r', label='$x+3y = 5$')
ax.plot(x, y2, color='g', label='$x-2y = -5$')
ax.plot(x, y3, color='b', label='$x+y = 3$')

# Plot the intersection points
for point in intersection:
    ax.plot(point[0], point[1], 'ko')
    ax.text(point[0], point[1], f"({point[0]:.2f}, {point[1]:.2f})", fontsize=9, ha='right')

# Set limits and properties
ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)
ax.set_xticks(np.arange(-4, 5))
ax.set_aspect('equal')
ax.grid(True, ls=':')
plt.xlabel("X")
plt.ylabel("Y")
ax.legend()
plt.show()
```

3. Create a 4x4 matrix containing random numbers in the range of 1 to 20. Have the user
prompted for a number, check that the imputed value is a number and then check if the
number is in the matrix. If the number is in the matrix, return the number of times that
number is in the matrix.

```python
import numpy as np


matrix = np.random.randint(1, 21, size=(4, 4))


while True:
    try:
        user_input = int(input("Enter a number to search in the matrix: "))
        break
    except ValueError:
        print("Please enter a valid integer.")


if user_input in matrix:
    count = np.count_nonzero(matrix == user_input)
    print(f"The number {user_input} is in the matrix {count} times.")
else:
    print("The number is not in the matrix.")
```

**Numpy.linalg**

1. linalg.det(a): calculates the determinant of the matrix

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
det_a = np.linalg.det(a)
print(det_a)
```

linalg.matrix_rank(a): computes the rank of the matrix

```python
import numpy as np

a = np.array([[1, 2], [3, 4]])
rank_a = np.linalg.matrix_rank(a)
print(rank_a)
```

3. linalg.inv(a): calculates the multiplicative inverse of a matrix

```python
import numpy as np
a_inv=np.linalg.inv(a)
print(a_inv )
```

4. linalg.solve(a,b): solve a linear system of equations ax=b

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([1,2])
x=np.linalg.solve(a,b)
print(x)
```

5. linalg.pinv(a): compute the pseudo-inverse of a matrix.

```python
import numpy as np

a = np.random.randn(3, 2)
b = np.linalg.pinv(a)

result = np.allclose(a, np.dot(a, np.dot(b, a)))
print(result)
```
